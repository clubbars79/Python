```python
data= pd.crosstab(df['price'], df['floors']).reset_index().sort_values(by='price', ascending=False)
data.rename(columns={1:'no', 2:'yes'}, inplace=True)
data.head()
```

```python
plt.figure(figsize=(30,8))
n_ticks = np.arange(len(data['yes']))
# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 2

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data['no'], width=w)
plt.bar(n_ticks + offset, data['yes'], width=w)

plt.title('Распределение стоимости домов относительно 1-х и 2-х этажных')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во домов')
plt.legend(['одноэтажные', 'двухэтажные'])
plt.xticks(n_ticks, data['price'], rotation = 45);  # добавляем метки делений
```

Вывод: При стоимости дома выше среднего, большенство домов в два этажа. В сегменте ниже среднего по стоимости, большинство одноэтажных домов
